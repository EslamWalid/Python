## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(1,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y,label='viz1')
plt.plot(x+2,y,'r+',color='blue',label='viz2')#cmykrgb
plt.plot(x+3,y,'g--',label='viz3')
plt.legend()
```

```python
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
d=[1,2,3,1,4,3,5,6,1,3,9,10]
plt.hist(d)
```
